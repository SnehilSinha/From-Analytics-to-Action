```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
#"How Data Has Transformed Film Industry Decision-Making"

# aspects of filmmaking
categories = ['Script Selection', 'Casting', 'Budget Allocation', 
              'Cinematography', "Editing","Sound Design", 'Marketing', 'Release Strategy']

#  level of datafication impact (0-10 scale)
impact_levels = [6,9,8,5,7,6,10,9]

# Create the horizontal bar chart
fig, ax = plt.subplots(figsize=(10, 6))
y_pos = np.arange(len(categories))

# Create bars with a color gradient based on impact level
bars = ax.barh(y_pos, impact_levels, align='center', 
               color=plt.cm.viridis(np.array(impact_levels)/10))

# Add labels and formatting
ax.set_yticks(y_pos)
ax.set_yticklabels(categories)
ax.invert_yaxis()  # Labels read top-to-bottom
ax.set_xlabel('Level of Datafication Impact (0-10)')
ax.set_title('How Data Has Transformed Film Industry Decision-Making')

# Add a color bar to show the scale
sm = plt.cm.ScalarMappable(cmap=plt.cm.viridis, norm=plt.Normalize(0, 10))
sm.set_array([])
cbar = plt.colorbar(sm)
cbar.set_label('Impact Intensity: Human Judgment   Algorithm Dominance')

# Add value labels on the bars
for i, v in enumerate(impact_levels):
    ax.text(v + 0.2, i, str(v), va='center')

plt.tight_layout()
plt.show()
```

*Importance of this data:*
This data mapping the degree of datafication across different aspects of filmmaking is essential for understanding how creative decisions are increasingly influenced or constrained by analytics in the modern film industry. By quantifying where data is most dominantsuch as in casting, marketing, and distributionwe highlight the tension between creative autonomy and commercial optimization, which is central to Will & Agencys mission to explore the role of creative freedom in a data-driven world.

*Relevancy:*
This insight supports our goal of identifying where and how creative decision-makers still retain agency, and where data may be overly dictating outcomes. It helps categorize which areas might benefit from rebalancing artistic and algorithmic inputs, thus directly feeding into Will & Agency's mission to protect and promote independent artistic voice. It fills a gap in showing not just that data is present, but where it exerts influenceand where artists might push back.
